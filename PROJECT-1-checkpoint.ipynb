 ORIGINAL RELEASE DATE AND SIZE :

      We will look at trends in app size over time.

            * First, Convert str to datetime type
            * re-index by using set_index

```python
df['Original Release Date'] = pd.to_datetime(df['Original Release Date'], format = '%d/%m/%Y')
date_size = pd.DataFrame({'size':df['Size']})
date_size = date_size.set_index(df['Original Release Date'])
date_size = date_size.sort_values(by=['Original Release Date'])
date_size.head()
```

   MONTH FOR TRENDS :
    
    The criteria for grouping depend on the parameters.
     We use resample method.

     * 'M'= 'end of month', 'Y'= 'Year'.
     * Finding size , minimum and maximum value.

```python
date_size['size'] = date_size['size'].apply(lambda b : b//(2**10))
monthly_size = date_size.resample('M').mean()
tmp = date_size.resample('M')
monthly_size['min'] = tmp.min()
monthly_size['max'] = tmp.max()
monthly_size.head()
```

  DATA VS APP SIZE(MONTHLY)

```python

fig = figure(x_axis_type='datetime',           
             plot_height=250, plot_width=750,
             title='Date vs App Size (Monthly)')
fig.line(y='size', x='Original Release Date', source=monthly_size, line_width=2, line_color='Green')
show(fig)
```

3) IDENTIFY TREND OF USER RATING BASED ON PRICING :

ARE PAID GAMES REALLY GOOD ?


```python
paid = df[df['Price']>0]
free = df[df['Price']==0]
fig, ax = plt.subplots(1, 2, figsize=(15,8))
sns.countplot(data=paid, y='Average User Rating', ax=ax[0], palette='plasma')
ax[0].set_title('Paid Games')
ax[0].set_xlim([0, 1000])

sns.countplot(data=free, y='Average User Rating', ax=ax[1], palette='viridis')
ax[1].set_title('Free Games')
ax[1].set_xlim([0,1000])
plt.tight_layout();
plt.show()
```

CONCLUSION :

  * As expected there are less number of paid games than free games
  * But Still we cannot see any difference in user ratings of both the categories
  * Most of the Games are rated quite good around 4.0-5.0
  * It doesn't seem like price has an impact on the ratings as both free and paid games have almost same ratings
